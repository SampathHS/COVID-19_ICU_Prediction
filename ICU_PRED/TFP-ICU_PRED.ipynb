```python
X = data.drop('ICU',axis=1)
y = data['ICU']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.3, random_state=0)

X_train.shape , X_test.shape
```

```python
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(X_train, 0.9)
len(set(corr_features))
```

```python
X_train = X_train.drop(corr_features,axis=1)
X_test = X_test.drop(corr_features,axis=1)
```

```python
X_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegressionCV
from sklearn.pipeline import make_pipeline
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
```

```python
def res(y_true,y_pred):
  tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
  accuracy=(tp+tn)/(tp+fp+fn+tn)
  specificity = tn/(tn+fp)
  sensitivity=tp/(tp+fn)
  print("Accuracy:",accuracy*100)
  print("Sensitivity:",sensitivity*100)
  print("Specificity:",specificity*100)
  print("ROC_AUC_Score:",roc_auc_score(y_true, y_pred)*100)
```

```python
lgc=make_pipeline(LogisticRegressionCV(cv=5,random_state=1,max_iter=100000))
lgc.fit(X_train, y_train)
y_pred=lgc.predict(X_test)
res(y_test,y_pred)
```

```python
from sklearn import svm
from sklearn.pipeline import make_pipeline
```

```python
SVM_object = make_pipeline(svm.SVC(kernel='linear'))
SVM_object.fit(X_train,y_train)
y_pred=SVM_object.predict(X_test)
res(y_test,y_pred)
```
